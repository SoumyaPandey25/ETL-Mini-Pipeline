```python
import pandas as pd

df = pd.read_csv('/content/Churn_Modelling.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#duplicate
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.dtypes
```

```python
df['creditscore'] = df['creditscore'].astype(int)
df['age'] = df['age'].astype(int)
```

```python
# age group
df['age_group'] = pd.cut(df['age'],
                         bins=[18,30,45,60,100],
                         labels=['young','adult','mid','senior'])

# balance flag
df['has_balance'] = df['balance'].apply(lambda x: 1 if x > 0 else 0)

# salary in k
df['salary_k'] = df['estimatedsalary'] / 1000
```

```python
df.to_csv('processed_data.csv', index=False)
```

```python
#Customer Table

customers = df[['customerid','surname','geography','gender','age']]
customers.to_csv('customers.csv', index=False)
```

```python
#Account Table

accounts = df[['customerid','creditscore','balance','numofproducts','hascrcard','isactivemember']]
accounts.to_csv('accounts.csv', index=False)
```

```python
#Churn Table

churn = df[['customerid','exited']]
churn.to_csv('churn.csv', index=False)
```

```python
import sqlite3

conn = sqlite3.connect('database.sqlite')

customers.to_sql('customers', conn, if_exists='replace', index=False)
accounts.to_sql('accounts', conn, if_exists='replace', index=False)
churn.to_sql('churn', conn, if_exists='replace', index=False)

conn.close()
```
